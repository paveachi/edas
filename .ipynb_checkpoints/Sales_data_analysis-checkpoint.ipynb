```python
import pandas as pd
import os
```

#### Task 1: Merging 12 months of sales data into 1 CSV

```python
april_df = pd.read_csv('./Sales_Data/Sales_April_2019.csv')

files = [file for file in os.listdir('./Sales_Data')]

for file in files:
    print(file)
```

```python
all_months_data = pd.DataFrame()

for file in files:
    df1 = pd.read_csv("./Sales_Data/" + file)
    all_months_data = pd.concat([all_months_data,df1])
    
all_months_data.to_csv('all_data.csv', index=False)
```

```python
all_months_data
```

#### Read in DataFrame with all of the data

```python
all_data = pd.read_csv('all_data.csv')
```

### CLEANING THE DATA!!!

#### Dropping rows with NAN values 

```python
all_data = all_data.dropna()
```

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data.head()
```

##### Add additional columns to data

#### Task #2: adding a month column

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

#### Adding a Sales column

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
```

```python
all_data.head()
```

#### Adding a city column

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x : get_city(x) + ' ' + get_state(x))

all_data.head()
```

#### Question 1: What was the best month for sales? How much was earned that month?

```python
results = all_data.groupby('Month').sum()
results
```

```python
import matplotlib.pyplot as plt

months = range(1,13)

plt.bar(months, results.Sales)

plt.xticks(months)
plt.ylabel('Sales in USD($)')
plt.xlabel('Month Number')

plt.show()
```

It is clear that December was the best month for sales.

#### Question #2: What US city had the highest number of sales?

```python
city_sales = all_data.groupby('City').sum()
city_sales
```

```python
import matplotlib.pyplot as plt
import numpy as np

cities = all_data['City'].unique()
cities = np.sort(cities)

plt.bar(cities, city_sales.Sales)

plt.xticks(cities, rotation='vertical', size=8)
plt.ylabel('Sales in USD($)')
plt.xlabel('city name')

plt.show()
```

#### What time do people place orders?

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()
```

```python
hours = range(0,24)

plt.plot(hours, all_data.groupby('Hour').count())

plt.title('Orders By Hour')

plt.xticks(hours)
plt.xlabel('Hour of Day')
plt.ylabel('Number of Orders')

plt.grid()

plt.show()
```

#### What products are typically sold together? 

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df = df[['Order ID', 'Grouped']].drop_duplicates()

df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
    
count.most_common(5)
```

### What product sold the most?

```python
results = all_data.groupby('Product').sum()
results
```

```python
import numpy as np

product = all_data['Product'].unique()

product = np.sort(product)

quantity_sold = results['Quantity Ordered']

plt.bar(product, quantity_sold)

plt.xticks(product, rotation='vertical', size=8)
plt.yticks()
plt.ylabel('Quantity Ordered')
plt.xlabel('Product')

plt.show()
```

```python
prices = all_data.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(product, quantity_sold, color='g')
ax2.plot(product, prices, 'b-')


ax1.set_xlabel('Prodcut Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(product, rotation='vertical', size=8)

plt.show()
```
